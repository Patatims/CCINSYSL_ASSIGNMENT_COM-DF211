Build a Simple Linear Regression model using **Gradient Descent** to study the linear relationship between the years of experience and salary.

7. Build the model using gradient descent

```python
X= df['YearsExperience'].values
y= df['Salary'].values
```

```python
#Building the model
m= 0
b= 0

L = 0.00001 # L stands for Learning Rate
epochs = 1500 #The number of iterations  to perform gradient descent

n = float(len(X)) #The number of Elements in variable 'X'
```

8. Print the values of the slope and the intercept

- Display the slope
- Display the intercept

```python
#Performing the Gradient Descent
for i in range(epochs):
  y_pred = m * X + b #This is the current predicted Value of salary(Y)

  D_m = (-2/n) * sum (X* (y -y_pred)) #The Derivative wrt m
  D_b = (-2/n) * sum (y -y_pred) #The Derivative wrt c

  m = m - L * D_m #To update m

  b = b - L * D_b #To update b

print ("Slope: ", m)
print ("y-intercept: ", b)
```

9. Draw the Regression Line

```python
y_pred = m*X + b

plt.scatter(X, y, color = 'blue', label="Data Points")
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red', label = 'Regression Line')  # regression line
plt.title('Relationshp of YearExperience and Salary with Gradient Descent')
plt.xlabel('Years of Experience')
plt.ylabel('The Salary')
plt.legend(loc=4)
plt.show()
```

10. Compute the Sum of Square Error (SSE)

```python
df1 = pd.DataFrame({"Actual_Salary": y, "Predicted_Salary": y_pred})
df1.head()
```

```python
sum_of_squared_errors = str(np.sum(np.square(df1['Predicted_Salary'] - df1['Actual_Salary'])))

print('The Residuals sum of square is : ', sum_of_squared_errors)
```
